```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the dataset
dataset = sns.load_dataset("iris")
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,4].values

dataset.head()
```

```python
#label encoding the Target values as the values are Object type

from sklearn.preprocessing import LabelEncoder
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
mapping = {l: i for i, l in enumerate(labelencoder_y.classes_)}
mapping
```

```python
#Reducing the dimensions for 3D graph as the number of input features are 4

from sklearn.decomposition import PCA
pca = PCA(n_components=3)
X = pca.fit_transform(X)
X[:10]
```

```python
#Plotting and Visualizing the 3D chart

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(15, 10))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

plt.cla()

# Reorder the labels
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, s = 60, edgecolor='k')
plt.show()
```
